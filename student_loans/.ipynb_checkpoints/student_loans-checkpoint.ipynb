```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
loan_type_file = "raw_data/PortfoliobyLoanType.xls"
loan_type = pd.read_excel(loan_type_file)
#loan_type
#all dollars in billions
#all recipients in millions
```

```python
loan_type['Federal Fiscal Year'] = loan_type['Federal Fiscal Year'].astype(str)
```

```python
loan_type['axis_labels'] = loan_type['Federal Fiscal Year'] + "_" + loan_type['quarter']
```

```python
loan_type_dollars = loan_type.iloc[:, [6, 8, 10, 12, 14]]
```

```python
loan_type_recipients = loan_type.iloc[:, [7, 9, 11, 13, 15]]
```

```python
loan_type['total_dollars'] = loan_type_dollars.agg("sum", axis="columns")
loan_type['total_recipients'] = loan_type_recipients.agg("sum", axis="columns")
#loan_type
```

```python
x_axis = loan_type['axis_labels']
y_axis = loan_type['total_dollars']
y_recip = loan_type['total_recipients']
```

```python
#ratio of min y_lim to max y_lim
loan_plot_ylim_ratio = 1000/1500
#proportionalize recipient chart y_axis
recip_plot_max_ylim = 55
min_ylim = recip_plot_max_ylim*loan_plot_ylim_ratio
```

```python
plt.figure(figsize=(20,5))
fig, ax1 = plt.subplots()
ax1.bar(x_axis, y_axis)
ax1.set_ylim(1000, 1500)
ax2 = ax1.twinx()
ax2.plot(x_axis, y_recip, color='red')
ax2.set_ylim(min_ylim, recip_plot_max_ylim)
plt.title('Billions of Dollars of Loans vs. Millions of Loan Recipients')
fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 20
fig_size[1] = 8
```
