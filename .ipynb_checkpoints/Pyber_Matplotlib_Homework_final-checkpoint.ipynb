```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
from matplotlib import colors as mcolors
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"; ride_data_to_load = "data/ride_data.csv"
# Read the City and Ride Data
city_df = pd.read_csv(city_data_to_load); ride_df = pd.read_csv(ride_data_to_load) 
# Combine the data into a single dataset
Data = ride_df.merge(city_df, 'inner')
# Display the data table for preview
Data = Data[['city', 'date', 'fare', 'ride_id', 'driver_count', 'type']]
```

```python
temp = Data.drop_duplicates('city')
```

```python
total_faresum = Data['fare'].sum().round(2); 
print("Total Fare: $" + str(total_faresum))
total_ridecount = Data['ride_id'].count(); 
print("Total Rides: " + str(total_ridecount))
total_drivercount = temp['driver_count'].sum(); 
print("Total Drivers: " + str(total_drivercount))
```

```python
urban1 = (Data[Data['type'] == "Urban"]); urban_ride = urban1.groupby('city').count()
urban2 = urban1.drop(['ride_id'], axis=1).groupby('city').mean().round(2); urban2['ride_count'] = urban_ride['ride_id']
urban = urban2
```

```python
urban_faresum = urban1['fare'].sum().round(2); 
print("Urban Fare: $" + str(urban_faresum))
urban_ridecount = urban['ride_count'].sum(); 
print("Urban Rides: " + str(urban_ridecount))
urban_drivercount = (temp['driver_count'][temp['type'] == "Urban"]).sum(); 
print("Urban Drivers: " + str(urban_drivercount))
urban_farepct = ((urban_faresum/total_faresum)*100).round(1); 
print("Urban Fare pct: " + str(urban_farepct))
urban_ridepct = ((urban_ridecount/total_ridecount)*100).round(1); 
print("Urban Rides pct: " + str(urban_ridepct))
urban_driverpct = ((urban_drivercount/total_drivercount)*100).round(1); 
print("Urban Drivers pct: " + str(urban_driverpct))
```

```python
suburban1 = (Data[Data['type'] == "Suburban"]); suburban_ride = suburban1.groupby('city').count()
suburban2 = suburban1.drop(['ride_id'], axis=1).groupby('city').mean().round(2); suburban2['ride_count'] = suburban_ride['ride_id']
suburban = suburban2
```

```python
suburban_faresum = suburban1['fare'].sum().round(2); 
print("Subrban Fare: $" + str(suburban_faresum))
suburban_ridecount = suburban['ride_count'].sum(); 
print("Suburban Rides: " + str(suburban_ridecount))
suburban_drivercount = (temp['driver_count'][temp['type'] == "Suburban"]).sum(); 
print("Suburban Drivers: " + str(suburban_drivercount))
suburban_farepct = ((suburban_faresum/total_faresum)*100).round(1); 
print("Suburban Fare pct: " + str(suburban_farepct))
suburban_ridepct = ((suburban_ridecount/total_ridecount)*100).round(1); 
print("Suburban Rides pct: " + str(suburban_ridepct))
suburban_driverpct = ((suburban_drivercount/total_drivercount)*100).round(1); 
print("Suburban Drivers pct: " + str(suburban_driverpct))
```

```python
rural1 = (Data[Data['type'] == "Rural"]); rural_ride = rural1.groupby('city').count()
rural2 = rural1.drop(['ride_id'], axis=1).groupby('city').mean().round(2); rural2['ride_count'] = rural_ride['ride_id']
rural = rural2
```

```python
rural_faresum = rural1['fare'].sum().round(2); 
print("Rural Fare: $" + str(rural_faresum))
rural_ridecount = rural['ride_count'].sum(); 
print("Rural Rides: " + str(rural_ridecount))
rural_drivercount = (temp['driver_count'][temp['type'] == "Rural"]).sum(); 
print("Rural Drivers: " + str(rural_drivercount))
rural_farepct = ((rural_faresum/total_faresum)*100).round(1); 
print("Rural Fare pct: " + str(rural_farepct))
rural_ridepct = ((rural_ridecount/total_ridecount)*100).round(1); 
print("Rural Rides pct: " + str(rural_ridepct))
rural_driverpct = ((rural_drivercount/total_drivercount)*100).round(1); 
print("Rural Drivers pct: " + str(rural_driverpct))
```

```python
x_u = urban['ride_count']; y_u = urban['fare']; z_u = urban['driver_count']
```

```python
x_s = suburban['ride_count']; y_s = suburban['fare']; z_s = suburban['driver_count']
```

```python
x_r = rural['ride_count']; y_r = rural['fare']; z_r = rural['driver_count']
```

```python
plt.scatter(x_u, y_u, s=z_u*10, color = "lightcoral", linewidth=1, alpha=0.8, edgecolors="black", label='Urban')
plt.scatter(x_s, y_s, s=z_s*10, color = "lightskyblue", linewidth=1, alpha=0.8, edgecolors="black", label='Suburban')
plt.scatter(x_r, y_r, s=z_r*10, color = "gold", linewidth=1, alpha=0.8, edgecolors="black", label='Rural')
plt.xlabel("Total Number of Rides (per city)"); plt.ylabel("Average Fair ($)"); plt.ylim(15, 45)
plt.title("Pyber Ride Sharing Data (2016)"); plt.legend(loc='upper right'); plt.grid(b=None, which='major', axis='both')
plt.text(42, 40, "Note: "); plt.text(42, 38, "Circle size correlates with driver count per city.")
plt.show()
```

```python
fare_data = {'index': ['Rural fare', 'Suburban fare', 'Urban fare' ],
             'Fare data (in $)': [rural_faresum, suburban_faresum, urban_faresum]}
city_fare = pd.DataFrame(fare_data); city_fare = city_fare.set_index('index')
city_fare['Fare pct'] = ((city_fare['Fare data (in $)']/total_faresum)*100).round(2)
city_fare
```

```python
labels = ['Rural', 'Suburban', 'Urban']; explode = (0.1, 0.1, 0.1); colors = [ "gold", "lightskyblue", "lightcoral"]
fig1, ax1 = plt.subplots()
ax1.pie(city_fare['Fare pct'], explode=explode, colors=colors, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=120)
ax1.axis('equal'); plt.title("% of Total Fare by City Type"); plt.show()
```

```python
ride_data = {'index': ['Rural Rides', 'Suburban Rides', 'Urban Rides'],
             'Ride Data': [rural_ridecount, suburban_ridecount, urban_ridecount]}
city_rides = pd.DataFrame(ride_data); city_rides = city_rides.set_index('index')
city_rides['Ride pct'] = ((city_rides['Ride Data']/total_ridecount)*100).round(2)
city_rides
```

```python
labels = ['Rural', 'Suburban', 'Urban']; explode = (0.1, 0.1, 0.1); colors = ["gold", "lightskyblue", "lightcoral"]
fig1, ax1 = plt.subplots()
ax1.pie(city_rides['Ride pct'], explode=explode, colors=colors, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=120)
ax1.axis('equal'); plt.title("% of Total Rides by City Type"); plt.show()
```

```python
driver_data = {'index': ['Rural Drivers', 'Suburban Drivers', 'Urban Drivers'],
               'Driver Data': [rural_drivercount, suburban_drivercount, urban_drivercount]}
city_drivers = pd.DataFrame(driver_data); city_drivers = city_drivers.set_index('index')
city_drivers['Driver pct'] = ((city_drivers['Driver Data']/total_drivercount)*100).round(2)
city_drivers
```

```python
labels = ['Rural', 'Suburban', 'Urban']; explode = (0.1, 0.1, 0.1); colors = ["gold", "lightskyblue", "lightcoral"]
fig1, ax1 = plt.subplots()
ax1.pie(city_drivers['Driver pct'], explode=explode, colors=colors, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=0)
ax1.axis('equal'); plt.title("% of Total Drivers by City Type"); plt.show()
```
